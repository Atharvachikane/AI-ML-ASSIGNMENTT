#***PROBLEM STAEMENT***

Problem Statement: Predict the salary of an employee based on their years of experience.

# **1. Lack of Additional Features (Limited Predictors)**

**Issue:**

The dataset only has one feature (YearsExperience), but salary depends on multiple factors such as education, job role, industry, location, and skills.
A model trained on just years of experience might not generalize well.

**Solution:**

Use a more detailed dataset with extra features like Education Level, Job Title, Industry, Location, and Skills.
If additional data isnâ€™t available, apply feature engineering, such as categorizing experience levels (e.g., Junior, Mid, Senior).


# **2. Non-linearity in Salary Growth**

**Issue:**

The salary does not increase linearly with experience.
In reality, early experience may have a steeper growth rate, while later years show slower growth.
A simple Linear Regression model may not capture this pattern well.

**Solution:**

Use Polynomial Regression or Non-linear models like Decision Trees or Random Forests to capture better salary growth patterns.
Transform features (e.g., YearsExperience^2) to allow for non-linearity.

# **4. Outliers in Salary Data**

**Issue:**

Some individuals might have very high or low salaries compared to their experience level.
This could be due to industry differences, job roles, or negotiation skills.
Outliers can affect the regression line and prediction accuracy.

**Solution:**

Detect and remove outliers using methods like Z-score or IQR (Interquartile Range).
Use Robust Regression techniques that are less sensitive to extreme values
